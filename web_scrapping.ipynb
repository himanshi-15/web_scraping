```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

import os.path
from os import path


categories=["Fruits & Vegetables","Foodgrains, Oil & Masala","Bakery, Cakes & Dairy","Beverages","Snacks & Branded Foods","Eggs, Meat & Fish","Gourmet & World Food"]

driver=webdriver.Chrome()
driver.get('https://www.bigbasket.com/')

for pro in categories:
    
    try:
        element=WebDriverWait(driver,10).until(
            EC.presence_of_element_located((By.LINK_TEXT,"SHOP BY CATEGORY"))
        )
        element.click()
        element=WebDriverWait(driver,10).until(
            EC.presence_of_element_located((By.LINK_TEXT,pro))
        )
        element.click()
        element=WebDriverWait(driver,10).until(
            EC.presence_of_element_located((By.LINK_TEXT,"ALL PRODUCTS"))
        )
        element.click()
    except:
        driver.quit()

    #search=driver.find_element_by_xpath('//*[@id="dynamicDirective"]/product-deck/section/div[2]/div[4]/div[1]/div/div/div[2]')

    item_list=driver.find_element_by_class_name('items')
    product_info_list = item_list.find_elements_by_class_name('item')


    product_list=[]

    for i in range(6,len(product_info_list)):
        if(product_info_list[i].text==' '):
            continue
        product=product_info_list[i]
        image=product.find_elements_by_tag_name('img')[-1].get_property('src')
        details = product.text.split('\n')
        price = details[details.index('MRP:')+1]
        temp = {'brand': details[1],
            'product name': details[2],
            'quantity': details[3],
            'price': price[price.rindex('Rs'):],
            'expected delivery': details[-3],
            'image': image,
            'category':pro}
        product_list.append(temp)


    alldetails=[]
    for i in product_list:
        alldetails.append(i)
        print(alldetails)
    
    df=pd.DataFrame(alldetails)
    
    if(path.exists("data.csv")):
        f = open('data.csv', 'a')
        if(pro == 'Fruits & Vegetables'):
            df.to_csv(f,index=False,line_terminator='\n')
        else:
            df.to_csv(f,header=False,index=False,line_terminator='\n')
        f.close()
    else:
        df.to_csv('data.csv',index=False,line_terminator='\n') 
    print(pro+' done \n')
    
    driver.back()
```
